### Creational patterns ‚Äì tvorba objekt≈Ø

#### Singleton
**Popis**: Zaji≈°≈•uje, ≈æe t≈ô√≠da m√° pouze jednu instanci a poskytuje glob√°ln√≠ p≈ô√≠stup k n√≠.

**Pou≈æit√≠**: Vhodn√© pro spr√°vu konfigurac√≠, logov√°n√≠ nebo p≈ôipojen√≠ k datab√°zi.  
**Pozor**: M≈Ø≈æe komplikovat unit testov√°n√≠ a zp≈Øsobovat probl√©my v paraleln√≠m programov√°n√≠.

```python
class Singleton:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance
```

#### Builder

**Pou≈æit√≠**: Vhodn√© pro konfiguraci objekt≈Ø s mnoha parametry.

**Pou≈æit√≠**: Nap≈ô. p≈ôi p≈ôizp≈Øsoben√≠ t≈ô√≠dy z√°vislosti na jin√© rozhran√≠.

#### Decorator
**Popis**: P≈ôid√°v√° nov√© funkce objektu bez √∫pravy jeho struktury.

**Pou≈æit√≠**: Vhodn√© pro dynamick√© roz≈°i≈ôov√°n√≠ funkcionalit (nap≈ô. middleware, logov√°n√≠).

### Behavioral patterns ‚Äì chov√°n√≠ objekt≈Ø

#### Observer
**Popis**: Umo≈æ≈àuje jednomu objektu sledovat zmƒõny jin√©ho objektu.

**Pou≈æit√≠**: Vhodn√© pro implementaci syst√©m≈Ø typu publish/subscribe, nap≈ô. v GUI aplikac√≠ch.

```python

    def update(self, message): pass

class Subject:
```

#### Strategy

### P≈ô√≠klad 5

### P≈ô√≠klad 6

P≈ô√≠klad ukazuje implementaci **Factory design patternu**, kter√Ω umo≈æ≈àuje vytv√°≈ôet objekty bez nutnosti specifikovat p≈ôesnou t≈ô√≠du, kterou m√° vytvo≈ôit.

### P≈ô√≠klad 7

### Kompletn√≠ k√≥d:

````
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        return "Haf!"

class Cat(Animal):
    def sound(self):
        return "Mnau!"

class AnimalFactory:
    @staticmethod
    def create_animal(animal_type):
        if animal_type == "dog":
            return Dog()
        elif animal_type == "cat":
            return Cat()
        else:
            raise ValueError("Nezn√°m√Ω typ zv√≠≈ôete")

# Testov√°n√≠ tov√°rny
factory = AnimalFactory()
dog = factory.create_animal("dog")
cat = factory.create_animal("cat")

print(dog.sound())  # Haf!
print(cat.sound())  # Mnau!
````

### P≈ô√≠klad 8

### Oƒçek√°van√Ω v√Ωstup:
```
Haf!
Mnau!
```

### P≈ô√≠klad 9

### Vysvƒõtlen√≠:
Tov√°rna (`AnimalFactory`) vytv√°≈ô√≠ konkr√©tn√≠ instance t≈ô√≠d dle zadan√©ho typu. Pou≈æit√≠ abstraktn√≠ t≈ô√≠dy `Animal` zajist√≠, ≈æe ka≈æd√° implementace mus√≠ m√≠t metodu `sound`.

# 3. **Observer - Pozorovatel**

### P≈ô√≠klad 10

### Co demonstruje:
Tento p≈ô√≠klad ukazuje implementaci **Observer patternu**, kter√Ω umo≈æ≈àuje sledovat zmƒõny v objektu a automaticky upozornit ostatn√≠ objekty.

### P≈ô√≠klad 11

### Kompletn√≠ k√≥d:

````
class Subject:
    def __init__(self):
        self._observers = []

    def attach(self, observer):
        self._observers.append(observer)

    def notify(self, message):
        for observer in self._observers:
            observer.update(message)

class Observer:
    def __init__(self, name):
        self.name = name

    def update(self, message):
        print(f"{self.name} obdr≈æel zpr√°vu: {message}")

# Testov√°n√≠
subject = Subject()
observer1 = Observer("Petr")
observer2 = Observer("Jana")

subject.attach(observer1)
subject.attach(observer2)

subject.notify("Ahoj v≈°ichni!")
````

### P≈ô√≠klad 12

### Oƒçek√°van√Ω v√Ωstup:
```
Petr obdr≈æel zpr√°vu: Ahoj v≈°ichni!
Jana obdr≈æel zpr√°vu: Ahoj v≈°ichni!
```

### P≈ô√≠klad 13

### Vysvƒõtlen√≠:
T≈ô√≠da `Subject` spravuje seznam pozorovatel≈Ø (`_observers`) a p≈ôi zmƒõnƒõ (vol√°n√≠ `notify`) upozorn√≠ v≈°echny. T≈ô√≠dy `Observer` implementuj√≠ metodu `update`, kter√° p≈ôij√≠m√° zpr√°vu.

# 4. **Repository - √ölo≈æi≈°tƒõ dat**

### P≈ô√≠klad 14

### Co demonstruje:
Uk√°zka implementace **Repository patternu**, kter√Ω oddƒõluje logiku p≈ô√≠stupu k dat≈Øm od aplikace.

### P≈ô√≠klad 15

### Kompletn√≠ k√≥d:

````
class UserRepository:
    def __init__(self):
        self._users = []

    def add_user(self, user):
        self._users.append(user)

    def find_by_id(self, user_id):
        for user in self._users:
            if user['id'] == user_id:
                return user
        return None

# Testov√°n√≠
repo = UserRepository()
repo.add_user({'id': 1, 'name': 'Petr'})
repo.add_user({'id': 2, 'name': 'Jana'})

user = repo.find_by_id(1)
print(user)  # {'id': 1, 'name': 'Petr'}
````

### P≈ô√≠klad 16

### Oƒçek√°van√Ω v√Ωstup:
```
{'id': 1, 'name': 'Petr'}
```

### P≈ô√≠klad 17

### Vysvƒõtlen√≠:
T≈ô√≠da `UserRepository` poskytuje rozhran√≠ pro p≈ôid√°n√≠ a vyhled√°n√≠ u≈æivatel≈Ø. D√≠ky tomu m≈Ø≈æe b√Ωt logika p≈ô√≠stupu k dat≈Øm snadno zmƒõnƒõna (nap≈ô. z pamƒõti na datab√°zi) bez ovlivnƒõn√≠ ostatn√≠ho k√≥du.
