````python
# ğŸ–¼ï¸ GRADIO - PrÃ¡ce s obrÃ¡zky

import gradio as gr
import numpy as np
from PIL import Image

def analyzuj_obrazek(image):
    """
    Simulace analÃ½zy obrÃ¡zku.
    V reÃ¡lnÃ© aplikaci zde bude volÃ¡nÃ­ AI modelu.
    """
    if image is None:
        return "Nahrajte obrÃ¡zek!"
    
    # ZÃ­skÃ¡me informace o obrÃ¡zku
    width, height = image.size
    
    # Simulace predikce (v reÃ¡lu by tu byl model)
    vysledky = {
        "KÃ¡men": np.random.random(),
        "NÅ¯Å¾ky": np.random.random(),
        "PapÃ­r": np.random.random()
    }
    
    # Normalizace na souÄet 1
    total = sum(vysledky.values())
    vysledky = {k: v/total for k, v in vysledky.items()}
    
    return vysledky

# VytvoÅ™enÃ­ rozhranÃ­ pro obrÃ¡zky
demo_image = gr.Interface(
    fn=analyzuj_obrazek,
    inputs=gr.Image(type="pil", label="Nahrajte obrÃ¡zek ruky"),
    outputs=gr.Label(num_top_classes=3, label="VÃ½sledek"),
    title="âœ‹ Detektor KÃ¡men/NÅ¯Å¾ky/PapÃ­r (Demo)",
    description="Toto je simulace. V dalÅ¡Ã­ ÄÃ¡sti pÅ™ipojÃ­me skuteÄnÃ½ model!",
    examples=[]  # Zde mÅ¯Å¾ete pÅ™idat vzorovÃ© obrÃ¡zky
)

demo_image.launch(share=False)
````

## ğŸ“– ÄŒÃ¡st 4: KompletnÃ­ aplikace s Teachable Machine

NynÃ­ spojÃ­me vÅ¡e dohromady - pÅ™ipojÃ­me nÃ¡Å¡ natrÃ©novanÃ½ model z Teachable Machine!

````python
# ğŸš€ KOMPLETNÃ APLIKACE - KÃ¡men, NÅ¯Å¾ky, PapÃ­r

import gradio as gr
import tensorflow as tf
import numpy as np
from PIL import Image
import urllib.request
import json
import os

# =====================================================
# ğŸ”§ KONFIGURACE - ZDE VLOÅ½TE VÃÅ  MODEL URL!
# =====================================================

# PÅ™Ã­klad URL z Teachable Machine:
# MODEL_URL = "https://teachablemachine.withgoogle.com/models/ABC123/"

# Pro demonstraci pouÅ¾ijeme simulovanÃ½ model
USE_REAL_MODEL = False  # ZmÄ›Åˆte na True, kdyÅ¾ mÃ¡te svÅ¯j model
MODEL_URL = ""  # Sem vloÅ¾te vÃ¡Å¡ URL

# =====================================================

# NÃ¡zvy tÅ™Ã­d
LABELS = ["KÃ¡men", "NÅ¯Å¾ky", "PapÃ­r"]

# NaÄtenÃ­ modelu (pokud je k dispozici)
model = None

if USE_REAL_MODEL and MODEL_URL:
    try:
        # StaÅ¾enÃ­ modelu
        model_path = tf.keras.utils.get_file(
            'model.h5',
            MODEL_URL + 'model.json',
            cache_dir='.',
            cache_subdir='tm_model'
        )
        model = tf.keras.models.load_model(model_path)
        print("âœ… Model naÄten!")
    except Exception as e:
        print(f"âš ï¸ NepodaÅ™ilo se naÄÃ­st model: {e}")
        print("   PouÅ¾Ã­vÃ¡m simulovanÃ½ model.")

def predict(image):
    """
    HlavnÃ­ predikÄnÃ­ funkce.
    
    Args:
        image: PIL Image objekt
    
    Returns:
        SlovnÃ­k {tÅ™Ã­da: pravdÄ›podobnost}
    """
    if image is None:
        return {label: 0.0 for label in LABELS}
    
    # PÅ™edzpracovÃ¡nÃ­ obrÃ¡zku
    image = image.resize((224, 224))  # Velikost pro Teachable Machine
    image_array = np.asarray(image)
    
    # Normalizace (Teachable Machine oÄekÃ¡vÃ¡ hodnoty -1 aÅ¾ 1)
    normalized = (image_array.astype(np.float32) / 127.0) - 1
    
    # PÅ™idÃ¡nÃ­ batch dimenze
    data = np.expand_dims(normalized, axis=0)
    
    if model is not None:
        # SkuteÄnÃ¡ predikce
        prediction = model.predict(data, verbose=0)
        probabilities = prediction[0]
    else:
        # SimulovanÃ¡ predikce (pro demonstraci)
        # V reÃ¡lu nahradÃ­me skuteÄnÃ½m modelem
        probabilities = np.random.dirichlet([2, 2, 2])  # NÃ¡hodnÃ© pravdÄ›podobnosti
    
    # VytvoÅ™enÃ­ vÃ½sledku
    results = {LABELS[i]: float(probabilities[i]) for i in range(len(LABELS))}
    
    return results

print("âœ… PredikÄnÃ­ funkce pÅ™ipravena!")
````

````python
# ğŸŒ SPUÅ TÄšNÃ WEBOVÃ‰ APLIKACE

# VytvoÅ™enÃ­ Gradio rozhranÃ­
app = gr.Interface(
    fn=predict,
    inputs=gr.Image(
        type="pil", 
        label="ğŸ“¸ Nahrajte obrÃ¡zek vaÅ¡Ã­ ruky",
        sources=["upload", "webcam"]  # UmoÅ¾nÃ­ nahrÃ¡nÃ­ i webkameru
    ),
    outputs=gr.Label(
        num_top_classes=3, 
        label="ğŸ¯ VÃ½sledek"
    ),
    title="âœ‹ KÃ¡men, NÅ¯Å¾ky, PapÃ­r - AI Detektor",
    description="""
    ### Jak pouÅ¾Ã­vat:
    1. Nahrajte obrÃ¡zek vaÅ¡Ã­ ruky nebo pouÅ¾ijte webkameru
    2. UkaÅ¾te gesto: âœŠ KÃ¡men, âœŒï¸ NÅ¯Å¾ky, nebo âœ‹ PapÃ­r
    3. AI model rozpoznÃ¡ vaÅ¡e gesto!
    
    *Model byl natrÃ©novÃ¡n pomocÃ­ Teachable Machine.*
    """,
    article="""
    ### O projektu
    Tento projekt je souÄÃ¡stÃ­ kurzu **AI pro kaÅ¾dÃ©ho**. 
    NauÄili jste se, jak vytvoÅ™it vlastnÃ­ AI klasifikÃ¡tor obrÃ¡zkÅ¯ 
    pomocÃ­ Teachable Machine a Gradio.
    """,
    theme=gr.themes.Soft()
)

# SpuÅ¡tÄ›nÃ­ aplikace
print("ğŸš€ SpouÅ¡tÃ­m aplikaci...")
print("\n" + "="*50)
print("Pokud vidÃ­te URL, otevÅ™ete ho v prohlÃ­Å¾eÄi!")
print("="*50 + "\n")

app.launch(share=True)  # share=True vytvoÅ™Ã­ veÅ™ejnÃ½ odkaz
````

## ğŸ“– ÄŒÃ¡st 5: RozÅ¡Ã­Å™enÃ­ projektu

### MoÅ¾nÃ¡ vylepÅ¡enÃ­:

1. **PÅ™idat vÃ­ce tÅ™Ã­d** - palec nahoru, OK gesto, atd.
2. **VylepÅ¡it UI** - pÅ™idat instrukce, animace
3. **HrÃ¡t proti AI** - implementovat logiku hry
4. **Statistiky** - sledovat ÃºspÄ›Å¡nost

````python
# ğŸ® BONUS: Hra proti AI!

import random

def hraj_proti_ai(vas_tah_obr):
    """
    Hra KÃ¡men, NÅ¯Å¾ky, PapÃ­r proti AI.
    
    Args:
        vas_tah_obr: PIL Image vaÅ¡eho gesta
    
    Returns:
        SlovnÃ­k s vÃ½sledky hry
    """
    if vas_tah_obr is None:
        return "Nahrajte obrÃ¡zek!", "â“", "ÄŒekÃ¡m na vÃ¡Å¡ tah..."
    
    # Detekce vaÅ¡eho tahu (pouÅ¾ijeme predict funkci)
    predikce = predict(vas_tah_obr)
    vas_tah = max(predikce, key=predikce.get)
    vas_tah_jistota = predikce[vas_tah]
    
    # AI volÃ­ nÃ¡hodnÄ›
    ai_tah = random.choice(["KÃ¡men", "NÅ¯Å¾ky", "PapÃ­r"])
    
    # Emoji pro tahy
    emoji = {"KÃ¡men": "âœŠ", "NÅ¯Å¾ky": "âœŒï¸", "PapÃ­r": "âœ‹"}
    
    # UrÄenÃ­ vÃ­tÄ›ze
    vyhry = {
        ("KÃ¡men", "NÅ¯Å¾ky"): "Vy",
        ("NÅ¯Å¾ky", "PapÃ­r"): "Vy",
        ("PapÃ­r", "KÃ¡men"): "Vy",
        ("NÅ¯Å¾ky", "KÃ¡men"): "AI",
        ("PapÃ­r", "NÅ¯Å¾ky"): "AI",
        ("KÃ¡men", "PapÃ­r"): "AI"
    }
    
    if vas_tah == ai_tah:
        vysledek = "ğŸ¤ REMÃZA!"
    elif vyhry.get((vas_tah, ai_tah)) == "Vy":
        vysledek = "ğŸ‰ VYHRÃLI JSTE!"
    else:
        vysledek = "ğŸ¤– AI VYHRÃLA!"
    
    # VÃ½stup
    vas_tah_text = f"{emoji[vas_tah]} {vas_tah} (jistota: {vas_tah_jistota:.1%})"
    ai_tah_text = f"{emoji[ai_tah]} {ai_tah}"
    
    return vas_tah_text, ai_tah_text, vysledek

# VytvoÅ™enÃ­ hernÃ­ho rozhranÃ­
hra = gr.Interface(
    fn=hraj_proti_ai,
    inputs=gr.Image(type="pil", label="ğŸ“¸ VÃ¡Å¡ tah", sources=["upload", "webcam"]),
    outputs=[
        gr.Textbox(label="ğŸ‘¤ VÃ¡Å¡ tah"),
        gr.Textbox(label="ğŸ¤– AI tah"),
        gr.Textbox(label="ğŸ† VÃ½sledek")
    ],
    title="ğŸ® KÃ¡men, NÅ¯Å¾ky, PapÃ­r - Hra proti AI!",
    description="UkaÅ¾te gesto a hrajte proti poÄÃ­taÄi!"
)

hra.launch(share=False)
````
