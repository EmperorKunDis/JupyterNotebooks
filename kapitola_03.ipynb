## ƒå√°st 3: Anal√Ωza sentimentu

Nejprve vyzkou≈°√≠me jednoduch√Ω √∫kol - **anal√Ωzu sentimentu** (pozitivn√≠/negativn√≠).

Toto je p≈ô√≠klad **√∫zk√© AI** - model je specializovan√Ω na jeden konkr√©tn√≠ √∫kol.

```python
# ANAL√ùZA SENTIMENTU
# ==================

# Vytvo≈ô√≠me pipeline pro sentiment anal√Ωzu
# Pipeline = jednoduch√Ω zp≈Øsob, jak pou≈æ√≠t p≈ôedtr√©novan√Ω model
sentiment_analyzer = pipeline("sentiment-analysis")

print("Model pro anal√Ωzu sentimentu naƒçten!")
```

```python
# Testovac√≠ texty (v angliƒçtinƒõ - model je anglick√Ω)
texty = [
    "I love this product! It's amazing!",
    "This is terrible. I hate it.",
    "The movie was okay, nothing special.",
    "Best day of my life!",
    "I'm disappointed with the service."
]

print("ANAL√ùZA SENTIMENTU")
print("=" * 50)

for text in texty:
    vysledek = sentiment_analyzer(text)[0]
    label = vysledek['label']
    score = vysledek['score']
    
    # Vizu√°ln√≠ indik√°tor
    emoji = "üòä" if label == "POSITIVE" else "üòû"
    
    print(f"{emoji} {label} ({score:.1%})")
    print(f"   '{text}'")
    print()
```

### Cviƒçen√≠: Zkuste vlastn√≠ texty

Upravte texty n√≠≈æe a spus≈•te anal√Ωzu:

```python
# CVIƒåEN√ç: Vlo≈æte vlastn√≠ texty (v angliƒçtinƒõ)
# ============================================

moje_texty = [
    "This is great!",      # Zmƒõ≈àte na vlastn√≠ text
    "I don't like this.",  # Zmƒõ≈àte na vlastn√≠ text
]

for text in moje_texty:
    vysledek = sentiment_analyzer(text)[0]
    print(f"{text}")
    print(f"  ‚Üí {vysledek['label']} ({vysledek['score']:.1%})")
    print()
```

## ƒå√°st 4: Generov√°n√≠ textu

Teƒè p≈ôejdeme ke **generativn√≠ AI** - model, kter√Ω vytv√°≈ô√≠ nov√Ω text.

Pou≈æijeme mal√Ω model GPT-2, kter√Ω bƒõ≈æ√≠ i na CPU v Colabu.

```python
# GENEROV√ÅN√ç TEXTU
# ================

# Vytvo≈ô√≠me pipeline pro generov√°n√≠ textu
# Pou≈æ√≠v√°me mal√Ω model GPT-2, kter√Ω zvl√°dne i CPU
text_generator = pipeline("text-generation", model="gpt2")

print("Model GPT-2 naƒçten!")
print("Tento model m√° 124 milion≈Ø parametr≈Ø.")
print("Pro srovn√°n√≠: GPT-4 m√° ~1.8 bilionu parametr≈Ø.")
```

```python
# Generov√°n√≠ textu z poƒç√°teƒçn√≠ho promptu
prompt = "Artificial intelligence will change the world by"

print("GENEROV√ÅN√ç TEXTU")
print("=" * 50)
print(f"Prompt: '{prompt}'")
print()

# Generujeme text
vysledek = text_generator(
    prompt,
    max_length=80,        # Maxim√°ln√≠ d√©lka v√Ωstupu
    num_return_sequences=1,  # Poƒçet variant
    temperature=0.7,      # Kreativita (0=konzervativn√≠, 1=kreativn√≠)
    do_sample=True        # Pou≈æ√≠t sampling pro variabilitu
)

print("Vygenerovan√Ω text:")
print("-" * 50)
print(vysledek[0]['generated_text'])
```

```python
# V√≠ce variant stejn√©ho promptu
# Uk√°zka, ≈æe AI generuje r≈Øzn√© v√Ωstupy

prompt = "The future of work is"

print(f"Prompt: '{prompt}'")
print("=" * 50)

# Generujeme 3 r≈Øzn√© varianty
for i in range(3):
    vysledek = text_generator(
        prompt,
        max_length=50,
        num_return_sequences=1,
        temperature=0.9,
        do_sample=True
    )
    print(f"\nVarianta {i+1}:")
    print(vysledek[0]['generated_text'])
```

### Cviƒçen√≠: Experimentujte s parametry

**Temperature** ovliv≈àuje kreativitu:
- `temperature=0.1` ‚Üí konzervativn√≠, p≈ôedv√≠dateln√Ω text
- `temperature=1.0` ‚Üí kreativn√≠, nep≈ôedv√≠dateln√Ω text

```python
# CVIƒåEN√ç: Porovnejte r≈Øzn√© teploty
# =================================

prompt = "In the year 2050, robots will"

print("POROVN√ÅN√ç TEMPERATURE")
print("=" * 50)

for temp in [0.1, 0.5, 1.0]:
    vysledek = text_generator(
        prompt,
        max_length=60,
        temperature=temp,
        do_sample=True
    )
    print(f"\nTemperature = {temp}:")
    print(vysledek[0]['generated_text'])
```

## ƒå√°st 5: Zodpov√≠d√°n√≠ ot√°zek (Question Answering)

Dal≈°√≠ schopnost AI - extrahovat odpovƒõdi z textu.

```python
# QUESTION ANSWERING
# ==================

qa_model = pipeline("question-answering")

print("Model pro zodpov√≠d√°n√≠ ot√°zek naƒçten!")
```

```python
# Kontext (text, ze kter√©ho AI extrahuje odpovƒõdi)
context = """
Artificial Intelligence (AI) was founded as an academic discipline in 1956 
at the Dartmouth Conference. The field was founded on the assumption that 
human intelligence can be precisely described and simulated by machines. 
Early AI research focused on symbolic reasoning and problem-solving. 
Geoffrey Hinton, Yann LeCun, and Yoshua Bengio are considered the 
'godfathers of deep learning' for their work on neural networks. 
In 2012, AlexNet won the ImageNet competition and sparked the deep learning revolution.
"""

# Ot√°zky
otazky = [
    "When was AI founded?",
    "Where was the Dartmouth Conference?",
    "Who are the godfathers of deep learning?",
    "What happened in 2012?",
]

print("ZODPOV√çD√ÅN√ç OT√ÅZEK")
print("=" * 50)
print("Kontext: (text o historii AI)")
print()

for otazka in otazky:
    odpoved = qa_model(question=otazka, context=context)
    print(f"Q: {otazka}")
    print(f"A: {odpoved['answer']} (confidence: {odpoved['score']:.1%})")
    print()
```

## ƒå√°st 6: Sumarizace textu

AI um√≠ shrnout dlouh√Ω text do nƒõkolika vƒõt.

```python
# SUMARIZACE
# ==========

summarizer = pipeline("summarization")

print("Model pro sumarizaci naƒçten!")
```

```python
# Dlouh√Ω text k sumarizaci
dlouhy_text = """
Artificial intelligence has transformed numerous industries over the past decade.
In healthcare, AI systems can now analyze medical images to detect diseases like 
cancer with accuracy comparable to or exceeding human doctors. In finance, 
AI algorithms process millions of transactions per second to detect fraud. 
Self-driving cars use AI to perceive their environment and make split-second 
decisions. Language models like GPT can write essays, code programs, and even 
compose poetry. However, these advances also raise concerns about job displacement,
privacy, and the potential misuse of AI technology. Experts debate whether AI will
ultimately benefit humanity or pose existential risks. What is certain is that AI
will continue to evolve and shape our future in profound ways.
"""

print("ORIGIN√ÅLN√ç TEXT:")
print("-" * 50)
print(dlouhy_text)
print()

# Sumarizace
summary = summarizer(dlouhy_text, max_length=60, min_length=20, do_sample=False)

print("SHRNUT√ç:")
print("-" * 50)
print(summary[0]['summary_text'])
```

## ƒå√°st 7: Dopad AI na spoleƒçnost

Vytvo≈ôme vizualizaci dopad≈Ø AI na r≈Øzn√© oblasti.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data o dopadu AI
oblasti = ['Zdravotnictv√≠', 'Finance', 'Doprava', 'Vzdƒõl√°v√°n√≠', 'V√Ωroba', 'Z√°bava']
dopad_2020 = [40, 60, 30, 25, 50, 70]
dopad_2030 = [80, 85, 75, 70, 85, 90]

x = np.arange(len(oblasti))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))

bars1 = ax.bar(x - width/2, dopad_2020, width, label='2020', color='#3498db')
bars2 = ax.bar(x + width/2, dopad_2030, width, label='2030 (odhad)', color='#e74c3c')

ax.set_ylabel('M√≠ra vyu≈æit√≠ AI (%)', fontsize=12)
ax.set_title('Dopad AI na r≈Øzn√© odvƒõtv√≠', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(oblasti, fontsize=10)
ax.legend()
ax.set_ylim(0, 100)

# P≈ôid√°n√≠ hodnot nad sloupce
for bar in bars1:
    height = bar.get_height()
    ax.annotate(f'{height}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=9)

for bar in bars2:
    height = bar.get_height()
    ax.annotate(f'{height}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

print("\nPozn√°mka: ƒå√≠sla jsou ilustrativn√≠ odhady pro vzdƒõl√°vac√≠ √∫ƒçely.")
```
