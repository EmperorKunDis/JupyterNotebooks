## Cast 3: Prakticky projekt - Spam Detektor

Nyni spojime vse, co jsme se naucili, do jednoho praktickeho projektu.

**Cil:** Vytvorit AI system, ktery rozpozna spam emaily od bezne komunikace.

**Pouzijeme:**

- Naive Bayes klasifikator
- Text preprocessing
- Evaluace modelu

```python
# Krok 1: Import knihoven
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Knihovny naimportovany")
print("\nPouzite knihovny:")
print("   - pandas: prace s daty")
print("   - sklearn: strojove uceni")
print("   - CountVectorizer: prevod textu na cisla")
print("   - MultinomialNB: Naive Bayes klasifikator")
```

```python
# Krok 2: Vytvoreni trenovacich dat
# V realnem svete bychom meli tisice emailu, pro ukazku staci mensi dataset

data = {
    'text': [
        # SPAM emaily
        'Vyzvednete si svou vyhru zdarma!',
        'Exkluzivni nabidka jen pro vas, nevahejte!',
        'Kliknete zde pro neuveritelnou slevu!',
        'Ziskejte pujcku bez dolozeni prijmu, rychle a snadno.',
        'VYHRA! Gratulujeme, vyhrali jste iPhone!',
        'Zbohatnete pres noc s nasim systemem!',
        'Pilulky na hubnuti - 10 kg za tyden!',
        'Kliknete ihned a ziskejte bonus zdarma!',
        'Vyhodna nabidka konci dnes - nepromesknete!',
        'Ziskejte penize bez prace, kliknete zde!',
        'Super sleva 90% jen dnes! Nakupte hned!',
        'Vyhrali jste v loterii, kliknete pro vyber!',
        
        # HAM (normalni) emaily
        'Ahoj, jak se mas? Sejdeme se zitra?',
        'Zprava o stavu projektu je pripravena.',
        'Dekuji za vas email, odpovim co nejdrive.',
        'Potvrzeni vasi objednavky c. 12345.',
        'Zitra mame schuzku v 10:00.',
        'Posilam ti materialy ke studiu.',
        'Muzes mi prosim zavolat?',
        'Faktura za minuly mesic je v priloze.',
        'Pripominam zitrejsi deadline projektu.',
        'Diky za pomoc vcera, moc jsi mi pomohl.',
        'Schuzka byla presunuta na ctvrtek.',
        'Posli mi prosim adresu, kam to mam dorucit.'
    ],
    'kategorie': [
        # Labels (stitky)
        'spam', 'spam', 'spam', 'spam', 'spam', 'spam',
        'spam', 'spam', 'spam', 'spam', 'spam', 'spam',
        'ham', 'ham', 'ham', 'ham', 'ham', 'ham',
        'ham', 'ham', 'ham', 'ham', 'ham', 'ham'
    ]
}

df = pd.DataFrame(data)

print("TRENOVACI DATA PRO SPAM DETEKTOR")
print("=" * 50)
print(f"\nCelkem emailu: {len(df)}")
print(f"Spam: {len(df[df['kategorie'] == 'spam'])}")
print(f"Ham (normalni): {len(df[df['kategorie'] == 'ham'])}")

print("\nUkazka dat:")
print(df.head(10).to_string())
```

```python
# Krok 3: Prevod textu na cisla (vektorizace)
# AI neumi pracovat primo s textem - musime ho prevest na cisla

print("PREVOD TEXTU NA CISLA")
print("=" * 50)

# CountVectorizer spocita, kolikrat se kazde slovo vyskytuje
vectorizer = CountVectorizer()

# Priklad: jak to funguje
priklad_texty = ["Ahoj jak se mas", "Ahoj vyhra zdarma"]
priklad_vektory = vectorizer.fit_transform(priklad_texty)

print("\nJak funguje CountVectorizer:")
print(f"\n   Vstup: {priklad_texty}")
print(f"\n   Slovnik: {vectorizer.get_feature_names_out()}")
print(f"\n   Matice cetnosti:")
print(f"   {priklad_vektory.toarray()}")

print("\n   Interpretace:")
print("   - Prvni text: 'Ahoj jak se mas' -> [1,1,1,0,1,0]")
print("   - Druhy text: 'Ahoj vyhra zdarma' -> [1,0,0,1,0,1]")
print("   - Kazde cislo = pocet vyskytu daneho slova")
```

```python
# Krok 4: Priprava dat pro trenovani

print("PRIPRAVA DAT PRO TRENOVANI")
print("=" * 50)

# Novy vectorizer pro nase data
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['text'])
y = df['kategorie']

print(f"\nVelikost slovniku: {len(vectorizer.get_feature_names_out())} unikatnich slov")
print(f"Pocet emailu: {X.shape[0]}")

# Rozdeleni na trenovaci a testovaci data (75% / 25%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.25, 
    random_state=42
)

print(f"\nRozdeleni dat:")
print(f"   Trenovaci: {X_train.shape[0]} emailu (75%)")
print(f"   Testovaci: {X_test.shape[0]} emailu (25%)")

print("\nProc rozdelujeme data?")
print("   Model trenujeme na trenovacich datech,")
print("   ale testujeme na datech, ktera nikdy nevidel.")
print("   Tak zjistime, jak dobre zobecnuje.")
```

```python
# Krok 5: Trenovani modelu

print("TRENOVANI NAIVE BAYES KLASIFIKATORU")
print("=" * 50)

# Vytvoreni a trenovani modelu
model = MultinomialNB()
model.fit(X_train, y_train)

print("\nModel natrenovan!")

print("\nCo je Naive Bayes?")
print("""
   Naive Bayes je jednoduchy, ale efektivni algoritmus zalozeny
   na Bayesove teoremu. Predpoklada, ze vsechny priznaky (slova)
   jsou na sobe nezavisle (proto 'naive' = naivni).
   
   Pro spam detection funguje takto:
   1. Spocita, jak casto se kazde slovo vyskytuje v spam vs ham
   2. Pro novy email spocita pravdepodobnost spam/ham
   3. Vybere kategorii s vyssi pravdepodobnosti
   
   Slova jako 'zdarma', 'vyhra', 'kliknete' -> pravdepodobne spam
   Slova jako 'schuzka', 'projekt', 'dekuji' -> pravdepodobne ham
""")
```

```python
# Krok 6: Evaluace modelu

print("EVALUACE MODELU")
print("=" * 50)

# Predikce na testovacich datech
predictions = model.predict(X_test)

# Presnost
accuracy = accuracy_score(y_test, predictions)
print(f"\nPresnost (Accuracy): {accuracy * 100:.1f}%")

# Detailni report
print("\nDetailni report:")
print(classification_report(y_test, predictions, target_names=['ham', 'spam']))

# Confusion matrix
cm = confusion_matrix(y_test, predictions, labels=['ham', 'spam'])
print("\nConfusion Matrix:")
print("               Predikce")
print("              ham  spam")
print(f"Skutecnost ham  {cm[0][0]:3d}   {cm[0][1]:3d}")
print(f"          spam  {cm[1][0]:3d}   {cm[1][1]:3d}")

print("\nInterpretace:")
print(f"   - Spravne klasifikovane ham: {cm[0][0]}")
print(f"   - Spravne klasifikovane spam: {cm[1][1]}")
print(f"   - Ham oznaceny jako spam (false positive): {cm[0][1]}")
print(f"   - Spam oznaceny jako ham (false negative): {cm[1][0]}")
```

```python
# Krok 7: Testovani na novych emailech!

print("TESTOVANI SPAM DETEKTORU")
print("=" * 50)

def klasifikuj_email(text):
    """Klasifikuje email jako spam nebo ham"""
    text_vector = vectorizer.transform([text])
    predikce = model.predict(text_vector)[0]
    pravdepodobnosti = model.predict_proba(text_vector)[0]
    
    # Index pro spam a ham
    classes = model.classes_
    ham_idx = list(classes).index('ham')
    spam_idx = list(classes).index('spam')
    
    return {
        'predikce': predikce,
        'ham_prob': pravdepodobnosti[ham_idx],
        'spam_prob': pravdepodobnosti[spam_idx]
    }

# Testovaci emaily
testovaci_emaily = [
    "Ahoj Petre, posles mi prosim ten report?",
    "VYHRA! Kliknete zde a ziskejte iPhone zdarma!",
    "Schuzka presunuta na 14:00, prijdes?",
    "Exkluzivni sleva 80% pouze dnes! Nakupte ihned!",
    "Dekuji za spolupraci, tesim se na dalsi projekt.",
    "Ziskejte pujcku bez zaruky, rychle a snadno!"
]

print("\nVysledky klasifikace:")
print("-" * 70)

for email in testovaci_emaily:
    vysledek = klasifikuj_email(email)
    
    # Oznaceni podle vysledku
    oznaceni = "[SPAM]" if vysledek['predikce'] == 'spam' else "[OK]"
    
    print(f"\n{oznaceni} '{email[:50]}...'" if len(email) > 50 else f"\n{oznaceni} '{email}'")
    print(f"   -> {vysledek['predikce'].upper()}")
    print(f"   -> Jistota: Ham {vysledek['ham_prob']*100:.1f}% | Spam {vysledek['spam_prob']*100:.1f}%")
```

```python
# Krok 8: Interaktivni testovani

print("INTERAKTIVNI SPAM DETEKTOR")
print("=" * 50)

def interaktivni_test():
    """Interaktivni testovani spam detektoru"""
    print("\nZadejte text emailu pro klasifikaci (nebo 'konec' pro ukonceni):")
    
    while True:
        print()
        email = input("Email: ").strip()
        
        if email.lower() == 'konec':
            print("\nDekuji za testovani!")
            break
        
        if not email:
            print("   Zadejte text emailu")
            continue
        
        vysledek = klasifikuj_email(email)
        
        if vysledek['predikce'] == 'spam':
            print(f"   [SPAM] ({vysledek['spam_prob']*100:.1f}% jistota)")
        else:
            print(f"   [OK] Normalni email ({vysledek['ham_prob']*100:.1f}% jistota)")

# Pro spusteni odkomentujte:
# interaktivni_test()
```

```python
# Nahled na Blok 2: Ukazka prohledavani

print("NAHLED NA BLOK 2: ALGORITMY A HLEDANI")
print("=" * 50)

# Jednoduchy priklad BFS (Breadth-First Search)
from collections import deque

# Graf mest (zjednodusena mapa)
mapa = {
    'Praha': ['Brno', 'Plzen', 'Liberec'],
    'Brno': ['Praha', 'Ostrava', 'Olomouc'],
    'Plzen': ['Praha', 'Karlovy Vary'],
    'Liberec': ['Praha'],
    'Ostrava': ['Brno', 'Olomouc'],
    'Olomouc': ['Brno', 'Ostrava'],
    'Karlovy Vary': ['Plzen']
}

def najdi_cestu(start, cil):
    """Najde nejkratsi cestu pomoci BFS"""
    fronta = deque([(start, [start])])
    navstivene = {start}
    
    while fronta:
        mesto, cesta = fronta.popleft()
        
        if mesto == cil:
            return cesta
        
        for soused in mapa.get(mesto, []):
            if soused not in navstivene:
                navstivene.add(soused)
                fronta.append((soused, cesta + [soused]))
    
    return None

# Ukazka
print("\nHledani cesty pomoci BFS (Breadth-First Search):")
print("\n   Mapa ceskych mest:")
for mesto, sousede in mapa.items():
    print(f"   {mesto} -> {', '.join(sousede)}")

start, cil = 'Karlovy Vary', 'Ostrava'
cesta = najdi_cestu(start, cil)

print(f"\n   Hledam cestu: {start} -> {cil}")
print(f"   Nalezena cesta: {' -> '.join(cesta)}")

print("\nV Bloku 2 se naucime efektivnejsi algoritmy jako A*!")
```
