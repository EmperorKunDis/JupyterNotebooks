## 2. Hlavn√≠ koncepty

### GraphQL vs. REST comparison

GraphQL a REST jsou dva r≈Øzn√© p≈ô√≠stupy k v√Ωvoji API. REST pou≈æ√≠v√° HTTP metody (GET, POST, PUT, DELETE) a pevnƒõ definovan√© endpointy. Ka≈æd√Ω endpoint vr√°t√≠ pevnƒõ danou strukturu dat ‚Äì tedy klient nem≈Ø≈æe ovlivnit, jak√° data dostane.

GraphQL naopak umo≈æ≈àuje klientovi zadat p≈ôesn√Ω dotaz, kter√Ω urƒçuje, jak√° data chce. Nap≈ô√≠klad:

```graphql

  user(id: "123") {
    name
    email
    posts {
      title
      content
    }
  }

```

V tomto p≈ô√≠padƒõ klient z√≠sk√° jen to, co specifikoval ‚Äì jm√©no, e-mail a p≈ô√≠spƒõvky u≈æivatele. V REST API bychom pravdƒõpodobnƒõ museli zavolat nƒõkolik endpoint≈Ø nebo dostat v√≠ce dat, ne≈æ pot≈ôebujeme.

### Schema definition language

GraphQL API je definov√°no pomoc√≠ Schema Definition Language (SDL). Tato jazykov√° specifikace umo≈æ≈àuje popsat typy dat, jejich vazby a operace, kter√© mohou b√Ωt provedeny.

P≈ô√≠klad jednoduch√©ho sch√©matu:


Tady se definuj√≠ typy `User` a `Post`, jejich vztahy a mo≈æn√© dotazy. Vlastnost `!` znamen√°, ≈æe pole je povinn√©.

### Queries, mutations, subscriptions

GraphQL rozli≈°uje t≈ôi hlavn√≠ typy operac√≠:

- **Queries** ‚Äì pou≈æ√≠vaj√≠ se pro ƒçten√≠ dat.
- **Mutations** ‚Äì pou≈æ√≠vaj√≠ se pro √∫pravu dat (vkl√°d√°n√≠, aktualizace, maz√°n√≠).
- **Subscriptions** ‚Äì umo≈æ≈àuj√≠ sledovat zmƒõny dat v re√°ln√©m ƒçase (nap≈ô. pomoc√≠ WebSocketu).
