# -*- coding: utf-8 -*-
"""

### P≈ô√≠klad 3

P≈ô√≠klad 1: Z√°kladn√≠ argument parser
Demonstruje z√°kladn√≠ pou≈æit√≠ argparse pro zpracov√°n√≠ p≈ô√≠kaz≈Ø CLI.

import argparse

def main():
    # Vytvo≈ôen√≠ parseru a definice argument≈Ø

    # P≈ôid√°n√≠ voliteln√©ho argumentu pro popis √∫kolu

### P≈ô√≠klad 7

### Co p≈ô√≠klad demonstruje:
Z√°kladn√≠ implementaci CRUD operac√≠ (Create, Read, Update, Delete) s ukl√°d√°n√≠m dat do JSON souboru.

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""

### P≈ô√≠klad 8

P≈ô√≠klad 2: Z√°kladn√≠ CRUD operace
Demonstruje z√°kladn√≠ CRUD operace s JSON persistence layer.
"""

import argparse
import json
import os
from datetime import datetime

# N√°zev souboru pro ukl√°d√°n√≠ √∫kol≈Ø
TODO_FILE = "todos.json"

def load_todos():
    """Naƒçte existuj√≠c√≠ √∫koly ze souboru"""
    if not os.path.exists(TODO_FILE):
        return []
    
    with open(TODO_FILE, 'r', encoding='utf-8') as f:
        return json.load(f)

def save_todos(todos):
    """Ulo≈æ√≠ √∫koly do souboru"""
    with open(TODO_FILE, 'w', encoding='utf-8') as f:
        json.dump(todos, f, ensure_ascii=False, indent=2)

def add_todo(description):
    """P≈ôid√° nov√Ω √∫kol"""
    todos = load_todos()
    new_todo = {
        "id": len(todos) + 1,
        "description": description,
        "created_at": datetime.now().isoformat(),
        "completed": False
    }
    todos.append(new_todo)
    save_todos(todos)
    print(f"√ökol p≈ôid√°n: {description}")

def list_todos():
    """Zobraz√≠ v≈°echny √∫koly"""
    todos = load_todos()
    if not todos:
        print("≈Ω√°dn√© √∫koly")
        return
    
    for todo in todos:
        status = "‚úì" if todo["completed"] else "‚óã"
        print(f"{todo['id']}. {status} {todo['description']}")

def delete_todo(todo_id):
    """Sma≈æe √∫kol podle ID"""
    todos = load_todos()
    filtered_todos = [todo for todo in todos if todo["id"] != todo_id]
    
    if len(filtered_todos) == len(todos):
        print(f"√ökol s ID {todo_id} nebyl nalezen")
        return
    
    save_todos(filtered_todos)
    print(f"√ökol s ID {todo_id} smaz√°n")

def main():
    parser = argparse.ArgumentParser(description='CLI Todo aplikace')
    parser.add_argument('action', choices=['add', 'list', 'delete'])
    parser.add_argument('--description', '-d')
    parser.add_argument('--id', '-i', type=int)
    
    args = parser.parse_args()
    
    if args.action == 'add' and args.description:
        add_todo(args.description)
    elif args.action == 'list':
        list_todos()
    elif args.action == 'delete' and args.id:
        delete_todo(args.id)
    else:
        print("Chybn√© pou≈æit√≠. Pou≈æijte --help pro n√°povƒõdu.")

if __name__ == "__main__":
    main()
```

    from colorama import init, Fore, Style

def save_todos(todos):
    with open(TODO_FILE, 'w', encoding='utf-8') as f:
        json.dump(todos, f, ensure_ascii=False, indent=2)

def add_todo(description):
    todos = load_todos()
    new_todo = {
        "id": len(todos) + 1,
        "description": description,
        "created_at": datetime.now().isoformat(),
        "completed": False

    todos.append(new_todo)
    save_todos(todos)
    
    # Barevn√Ω v√Ωstup p≈ôi p≈ôid√°n√≠ √∫kolu

    for todo in todos:
        status = f"{Fore.GREEN}‚úì{Style.RESET_ALL}" if todo["completed"] else f"{Fore.RED}‚óã{Style.RESET_ALL}"

    todos = load_todos()
    filtered_todos = [todo for todo in todos if todo["id"] != todo_id]
    
    if len(filtered_todos) == len(todos):
        print(f"{Fore.RED}√ökol s ID {todo_id} nebyl nalezen{Style.RESET_ALL}")
        return
    
    save_todos(filtered_todos)
