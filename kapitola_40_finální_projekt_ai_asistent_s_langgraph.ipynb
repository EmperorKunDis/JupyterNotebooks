Pou≈æit√≠ RAG v kombinaci s LangGraph:
- Umo≈æ≈àuje dynamick√© naƒç√≠t√°n√≠ informac√≠ podle pot≈ôeby.
- Zamezuje generov√°n√≠ nepravdiv√Ωch informac√≠.
- Vhodn√© nap≈ô. pro syst√©my, kde se data ƒçasto mƒõn√≠ (nap≈ô. v√Ωpoƒçty, novinky).

P≈ô√≠klad: Asistent m√° p≈ô√≠stup k datab√°zi dokument≈Ø o produktu. P≈ôi ot√°zce ‚ÄûJak√Ω je v√Ωkon tohoto modelu?‚Äú vyhled√° relevantn√≠ ƒç√°sti z dokumentace a odpov√≠ na z√°kladƒõ nich.

Tento p≈ô√≠klad ukazuje z√°kladn√≠ strukturu AI asistenta s vyu≈æit√≠m LangGraph. Demonstruje jednoduchou komunikaci mezi u≈æivatelem a asistentem bez jak√Ωchkoli roz≈°√≠≈ôen√Ωch funkc√≠.

### P≈ô√≠klad 3

### Kompletn√≠ k√≥d

````
from langgraph.graph import StateGraph, END
from typing import TypedDict, Annotated
import operator

# Definice stavu
class AgentState(TypedDict):
    messages: Annotated[list, operator.add]

# Definice uzlu (funkce)
def agent_node(state):
    """Jednoduch√Ω uzel asistenta"""
    # Simulace odpovƒõdi asistenta
    response = {
        "role": "assistant",
        "content": "Ahoj! Jsem tv≈Øj AI asistent. Jak ti mohu pomoci?"
    }
    return {"messages": [response]}

# Vytvo≈ôen√≠ grafu
def create_basic_graph():
    # Inicializace grafu
    workflow = StateGraph(AgentState)
    
    # P≈ôid√°n√≠ uzlu
    workflow.add_node("agent", agent_node)
    
    # Nastaven√≠ poƒç√°teƒçn√≠ho uzlu
    workflow.set_entry_point("agent")
    
    # Nastaven√≠ koncov√©ho uzlu
    workflow.add_edge("agent", END)
    
    # Kompilace grafu
    return workflow.compile()

# Spu≈°tƒõn√≠ p≈ô√≠kladu
if __name__ == "__main__":
    print("Vytv√°≈ô√≠m z√°kladn√≠ AI asistent...")
    
    # Vytvo≈ôen√≠ grafu
    graph = create_basic_graph()
    
    # Spu≈°tƒõn√≠ s jednoduch√Ωm vstupem
    initial_state = {"messages": [{"role": "user", "content": "Ahoj!"}]}
    
    # V√Ωstup
    print("V√Ωstup:")
    for output in graph.stream(initial_state):
        print(output)
````

## 2. Spr√°va v√≠ce krok≈Ø konverzace

### P≈ô√≠klad 6

### N√°zev p≈ô√≠kladu

### P≈ô√≠klad 7

### Co demonstruje
Tento p≈ô√≠klad ukazuje, jak spravovat v√≠ce krok≈Ø konverzace mezi u≈æivatelem a asistentem. Ka≈æd√Ω krok je zpracov√°n jako samostatn√Ω uzel v grafu.

### P≈ô√≠klad 8

### Kompletn√≠ k√≥d

````
from langgraph.graph import StateGraph, END
from typing import TypedDict, Annotated
import operator

# Definice stavu s histori√≠ konverzace
class ConversationState(TypedDict):
    messages: Annotated[list, operator.add]
    conversation_history: list

# Uzel pro zpracov√°n√≠ u≈æivatelsk√© zpr√°vy
def process_user_message(state):
    """Zpracuje u≈æivatelskou zpr√°vu a p≈ôid√° ji do historie"""
    user_message = state["messages"][-1]  # Posledn√≠ zpr√°va od u≈æivatele
    
    # Simulace anal√Ωzy zpr√°vy
    analysis = f"Analyzuji zpr√°vu: '{user_message['content']}'"
    
    return {
        "messages": [{"role": "assistant", "content": analysis}],
        "conversation_history": state["conversation_history"] + [user_message]
    }

# Uzel pro odpovƒõƒè
def generate_response(state):
    """Generuje odpovƒõƒè na z√°kladƒõ historie konverzace"""
    last_user_msg = state["messages"][-1]  # Posledn√≠ zpr√°va od u≈æivatele
    
    # Zalo≈æeno na p≈ôedchoz√≠ konverzaci
    if "pomoc" in last_user_msg["content"].lower():
        response = "R√°d ti pomohu! M≈Ø≈æe≈° mi ≈ô√≠ct, o co p≈ôesnƒõ jde?"
    elif "jak se m√°≈°" in last_user_msg["content"].lower():
        response = "Jsem skvƒõle, dƒõkuji za dotaz!"
    else:
        response = "Rozum√≠m tv√© zpr√°vƒõ. Mohu ti pomoct s nƒõƒç√≠m konkr√©tn√≠m?"
    
    return {
        "messages": [{"role": "assistant", "content": response}],
        "conversation_history": state["conversation_history"]
    }

# Vytvo≈ôen√≠ grafu v√≠ce krok≈Ø
def create_conversation_graph():
    workflow = StateGraph(ConversationState)
    
    # P≈ôid√°n√≠ uzl≈Ø
    workflow.add_node("process_user", process_user_message)
    workflow.add_node("generate_response", generate_response)
    
    # Nastaven√≠ toku
    workflow.set_entry_point("process_user")
    workflow.add_edge("process_user", "generate_response")
    workflow.add_edge("generate_response", END)
    
    return workflow.compile()

# Spu≈°tƒõn√≠ p≈ô√≠kladu
if __name__ == "__main__":
    print("Vytv√°≈ô√≠m graf v√≠ce krok≈Ø konverzace...")
    
    # Vytvo≈ôen√≠ grafu
    graph = create_conversation_graph()
    
    # Simulace v√≠ce zpr√°v
    initial_state = {
        "messages": [
            {"role": "user", "content": "Ahoj, jak se m√°≈°?"},
            {"role": "user", "content": "Pot≈ôebuji pomoct s programov√°n√≠m"}
        ],
        "conversation_history": []
    }
    
    print("V√Ωstup:")
    for output in graph.stream(initial_state):
        print(f"V√Ωstup: {output}")
````

    """T≈ô√≠da pro pr√°ci s LLM modelem"""

        """
        Inicializuje LLM model
        
        Args:
            api_key (str): API kl√≠ƒç pro OpenAI
            model_name (str): N√°zev modelu
            temperature (float): Teplota generov√°n√≠ (0-2)
        """

            raise ValueError("API kl√≠ƒç nem≈Ø≈æe b√Ωt pr√°zdn√Ω")
            
        self.client = OpenAI(api_key=api_key)
        self.model_name = model_name

    def generate_response(self, prompt: str, context: Dict[str, Any] = None) -> str:
        """
        Vygeneruje odpovƒõƒè z LLM modelu
        
        Args:
            prompt (str): Prompt pro generov√°n√≠
            context (Dict[str, Any]): Kontext pro generov√°n√≠
            
        Returns:
            str: Vygenerovan√° odpovƒõƒè
        """
        try:
            messages = [
                {"role": "system", "content": "Jsi u≈æiteƒçn√Ω asistent."},
                {"role": "user", "content": prompt}
            ]
            
            if context:
                # P≈ôid√°me kontext do zpr√°v
                for key, value in context.items():
                    messages.append({"role": "user", "content": f"{key}: {value}"})
            
            response = self.client.chat.completions.create(
                model=self.model_name,
                messages=messages,
                temperature=self.temperature,
                max_tokens=1000
            )
            
            answer = response.choices[0].message.content.strip()
            logger.info("Odpovƒõƒè vygenerov√°na √∫spƒõ≈°nƒõ")
            return answer
            
        except Exception as e:
