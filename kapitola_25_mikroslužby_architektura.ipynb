````
# service_boundaries.py
"""
Service boundaries - RozhranÃ­ sluÅ¾eb

Tento pÅ™Ã­klad ukazuje jak by mÄ›ly bÃ½t definovÃ¡ny hranice mezi sluÅ¾bami.
"""

from abc import ABC, abstractmethod
import json

# AbstraktnÃ­ rozhranÃ­ pro sluÅ¾bu uÅ¾ivatelÅ¯
class UserServiceInterface(ABC):
    @abstractmethod
    def get_user(self, user_id):
        pass
    
    @abstractmethod
    def create_user(self, user_data):
        pass

# KonkrÃ©tnÃ­ implementace sluÅ¾by uÅ¾ivatelÅ¯
class UserDatabaseService(UserServiceInterface):
    def __init__(self):
        self.users = {}
    
    def get_user(self, user_id):
        return self.users.get(user_id, None)
    
    def create_user(self, user_data):
        user_id = len(self.users) + 1
        self.users[user_id] = {
            "id": user_id,
            "name": user_data["name"],
            "email": user_data["email"]
        }
        return self.users[user_id]

# AbstraktnÃ­ rozhranÃ­ pro sluÅ¾bu objednÃ¡vek
class OrderServiceInterface(ABC):
    @abstractmethod
    def create_order(self, user_id, product):
        pass
    
    @abstractmethod
    def get_order(self, order_id):
        pass

# KonkrÃ©tnÃ­ implementace sluÅ¾by objednÃ¡vek
class OrderDatabaseService(OrderServiceInterface):
    def __init__(self):
        self.orders = {}
    
    def create_order(self, user_id, product):
        order_id = len(self.orders) + 1
        self.orders[order_id] = {
            "id": order_id,
            "user_id": user_id,
            "product": product,
            "status": "created"
        }
        return self.orders[order_id]
    
    def get_order(self, order_id):
        return self.orders.get(order_id, None)

# SluÅ¾ba pro komunikaci mezi sluÅ¾bami
class OrderProcessingService:
    def __init__(self, user_service: UserServiceInterface, order_service: OrderServiceInterface):
        self.user_service = user_service
        self.order_service = order_service
    
    def process_order(self, user_data, product):
        # VytvoÅ™enÃ­ uÅ¾ivatele
        created_user = self.user_service.create_user(user_data)
        
        # VytvoÅ™enÃ­ objednÃ¡vky
        order = self.order_service.create_order(created_user["id"], product)
        
        return {
            "user": created_user,
            "order": order,
            "message": f"UÅ¾ivatel {created_user['name']} objednal {product}"
        }

# Demonstrace
print("=== SERVICE BOUNDARIES ===")

# VytvoÅ™enÃ­ sluÅ¾eb
user_service = UserDatabaseService()
order_service = OrderDatabaseService()

# VytvoÅ™enÃ­ sluÅ¾by pro zpracovÃ¡nÃ­ objednÃ¡vek
processing_service = OrderProcessingService(user_service, order_service)

# ZpracovÃ¡nÃ­ objednÃ¡vky
result = processing_service.process_order(
    {"name": "Petr Svoboda", "email": "petr@example.com"},
    "smartfon"
)

print(json.dumps(result, indent=2))
````
