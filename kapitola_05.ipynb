## ÄŒÃ¡st 3: SÃ­la stroje - Rychlost a pÅ™esnost

### Ãškol: SpoÄÃ­tat prvoÄÃ­sla do 100 000

Kolik Äasu by vÃ¡m to zabralo? AI to zvlÃ¡dne za zlomek sekundy.

```python
# Demonstrace rychlosti stroje - hledÃ¡nÃ­ prvoÄÃ­sel
def je_prvocislo(n):
    """ZjistÃ­, zda je ÄÃ­slo prvoÄÃ­slo."""
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def najdi_prvocisla(limit):
    """Najde vÅ¡echna prvoÄÃ­sla do danÃ©ho limitu."""
    return [n for n in range(2, limit) if je_prvocislo(n)]

# MÄ›Å™enÃ­ Äasu
print("ğŸ¤– STROJ: PoÄÃ­tÃ¡m prvoÄÃ­sla do 100 000...\n")

start = time.time()
prvocisla = najdi_prvocisla(100000)
cas = time.time() - start

print(f"â±ï¸ ÄŒas vÃ½poÄtu: {cas:.3f} sekund")
print(f"ğŸ“Š Nalezeno prvoÄÃ­sel: {len(prvocisla)}")
print(f"\nğŸ”¢ PrvnÃ­ch 20 prvoÄÃ­sel: {prvocisla[:20]}")
print(f"ğŸ”¢ PoslednÃ­ch 5 prvoÄÃ­sel: {prvocisla[-5:]}")

print("\nğŸ’¡ VHLED: ÄŒlovÄ›ku by tento Ãºkol zabral hodiny nebo dny!")
```

```python
# Demonstrace konzistence stroje - opakovanÃ¡ pÅ™esnost
print("ğŸ¤– STROJ: Demonstrace dokonalÃ© konzistence\n")

# 10x spoÄÃ­tÃ¡me souÄet ÄÃ­sel - stroj dÃ¡ vÅ¾dy stejnÃ½ vÃ½sledek
cisla = list(range(1, 1001))
vysledky = []

for i in range(10):
    vysledek = sum(cisla)
    vysledky.append(vysledek)

print(f"SouÄet ÄÃ­sel 1-1000 (10 opakovÃ¡nÃ­): {vysledky}")
print(f"\nVÅ¡echny vÃ½sledky stejnÃ©: {len(set(vysledky)) == 1}")

print("\nğŸ’¡ VHLED: Stroj se nikdy neunavÃ­ a nedÄ›lÃ¡ chyby z nepozornosti!")
```

## ÄŒÃ¡st 4: SÃ­la stroje - RozpoznÃ¡vÃ¡nÃ­ vzorÅ¯ ve velkÃ½ch datech

AI dokÃ¡Å¾e analyzovat miliony datovÃ½ch bodÅ¯ a najÃ­t vzory, kterÃ© jsou pro ÄlovÄ›ka neviditelnÃ©.

```python
# Simulace analÃ½zy velkÃ½ch dat
print("ğŸ¤– STROJ: AnalÃ½za 10 000 transakcÃ­ pro detekci podvodu\n")

# Generujeme simulovanÃ¡ data
np.random.seed(42)
n_transakci = 10000

# NormÃ¡lnÃ­ transakce: ÄÃ¡stka 10-500 KÄ
# PodezÅ™elÃ© transakce: ÄÃ¡stka > 5000 KÄ nebo v noci (hodina 1-5)
transakce = []
for i in range(n_transakci):
    castka = np.random.exponential(100) + 10
    hodina = np.random.randint(0, 24)
    
    # OznaÄenÃ­ podezÅ™elÃ½ch
    podezrela = castka > 5000 or (hodina >= 1 and hodina <= 5 and castka > 1000)
    transakce.append({'castka': castka, 'hodina': hodina, 'podezrela': podezrela})

# AnalÃ½za
start = time.time()
podezrele = [t for t in transakce if t['podezrela']]
cas_analyzy = time.time() - start

print(f"ğŸ“Š Celkem transakcÃ­: {n_transakci}")
print(f"ğŸš¨ PodezÅ™elÃ½ch transakcÃ­: {len(podezrele)}")
print(f"â±ï¸ ÄŒas analÃ½zy: {cas_analyzy*1000:.2f} ms")

print("\nğŸ“ˆ PÅ™Ã­klady podezÅ™elÃ½ch transakcÃ­:")
for t in podezrele[:5]:
    print(f"   ÄŒÃ¡stka: {t['castka']:.2f} KÄ, Hodina: {t['hodina']}:00")

print("\nğŸ’¡ VHLED: ÄŒlovÄ›ku by trvalo hodiny projÃ­t 10 000 transakcÃ­!")
```
