## 2. Hlavn√≠ koncepty

### Pytest framework advanced

`Pytest` je jedna z nejroz≈°√≠≈ôenƒõj≈°√≠ch knihoven pro testov√°n√≠ v Pythonu. V pokroƒçil√© √∫rovni umo≈æ≈àuje vytv√°≈ôet komplexn√≠ testy s pou≈æit√≠m pokroƒçil√Ωch funkc√≠ jako jsou parametrizace, fixtures, a integrace s jin√Ωmi n√°stroji.

P≈ô√≠klad:
```python
import pytest

def test_calculate_discount():
    assert calculate_discount(100, 10) == 90
```

V pokroƒçil√©m pou≈æit√≠ lze vyu≈æ√≠t nap≈ô. **fixtures** ‚Äì funkce, kter√© p≈ôipravuj√≠ prost≈ôed√≠ pro testy:

```python
@pytest.fixture
def sample_data():
    return {"items": [1, 2, 3], "total": 6}
```

Tato fixture se pak m≈Ø≈æe pou≈æ√≠t v testech:
```python
def test_total_calculation(sample_data):
    assert sample_data["total"] == sum(sample_data["items"])
```

### Fixtures a parametrize

Fixtures jsou jedn√≠m z hlavn√≠ch n√°stroj≈Ø v `pytest`. Umo≈æ≈àuj√≠ p≈ôipravovat spoleƒçn√° data nebo prost≈ôed√≠ pro v√≠ce test≈Ø. M≈Ø≈æeme je pou≈æ√≠t jako **session**, **module**, nebo **function** scope.

```python
@pytest.fixture(scope="function")
def database_connection():
    # inicializace spojen√≠ s datab√°z√≠
    conn = create_connection()
    yield conn
    conn.close()
```

Funkce `parametrize` umo≈æ≈àuje spou≈°tƒõt stejn√Ω test s r≈Øzn√Ωmi vstupy:

```python
@pytest.mark.parametrize("input_value,expected", [
    (2, 4),
    (3, 9),
    (4, 16)
])
def test_square(input_value, expected):
    assert square(input_value) == expected
```

### Mocking a monkeypatching

Mocking je technika, kter√° nahrazuje skuteƒçn√© objekty nebo funkce testovac√≠mi "mock" objekty. V Pythonu se pro to pou≈æ√≠v√° nap≈ô. modul `unittest.mock`.

```python
from unittest.mock import Mock, patch

def test_api_call():
    mock_response = Mock()
    mock_response.json.return_value = {"status": "ok"}
    
    with patch("requests.get", return_value=mock_response):
        result = api_call()
        assert result["status"] == "ok"
```

Monkeypatching je alternativa k mockingu, kter√° umo≈æ≈àuje dynamickou zmƒõnu chov√°n√≠ objekt≈Ø p≈ôi bƒõhu testu:

```python
def test_function_with_mock(monkeypatch):
    monkeypatch.setattr("my_module.external_function", lambda: "mocked")
    assert my_module.my_function() == "mocked"
```

### Integration testing

Integraƒçn√≠ testov√°n√≠ ovƒõ≈ôuje, jak spr√°vnƒõ funguj√≠ komponenty mezi sebou. Nap≈ô√≠klad testuje, jak API komunikuje s datab√°z√≠ nebo zda v≈°echny moduly spolu pracuj√≠ podle oƒçek√°v√°n√≠.

P≈ô√≠klad:
```python
def test_user_creation_integration(db):
    user = create_user("John", "john@example.com")
    assert user.email == "john@example.com"
    assert db.session.query(User).filter_by(email="john@example.com").first() is not None
```

### End-to-end testing

End-to-end (E2E) testy simuluje re√°ln√© pou≈æit√≠ aplikace, od u≈æivatelsk√©ho rozhran√≠ a≈æ po backend. Vyu≈æ√≠vaj√≠ nap≈ô. n√°stroje jako `Selenium`, `Playwright` nebo `Cypress`.

```python
from selenium import webdriver

def test_login_page():
    driver = webdriver.Chrome()
    driver.get("http://localhost:8000/login")
    assert "Login" in driver.title
```

### Test coverage

Testovac√≠ pokryt√≠ (test coverage) mƒõ≈ô√≠, jak√° ƒç√°st k√≥du je p≈ôi testov√°n√≠ pokryta. N√°stroj `coverage.py` umo≈æ≈àuje z√≠skat podrobn√Ω report:

```bash
coverage run -m pytest tests/
coverage report
```

Pokud m√°me nap≈ô. 95 % pokryt√≠, znamen√° to, ≈æe 95 % k√≥du bylo spu≈°tƒõno bƒõhem test≈Ø.

### TDD a BDD approaches

Test-Driven Development (TDD) je metodologie, p≈ôi kter√© se nejprve nap√≠≈°e test, pot√© implementace, a pak se zase uprav√≠ k√≥d pro splnƒõn√≠ testu.

```python
def test_addition():
    assert add(2, 3) == 5  # Nejd≈ô√≠ve test, pak implementace
```

Behavior-Driven Development (BDD) zamƒõ≈ôuje testy na chov√°n√≠ aplikace z pohledu u≈æivatele a pou≈æ√≠v√° jazyk podobn√Ω p≈ôirozen√©mu, nap≈ô. `Given`, `When`, `Then`.

P≈ô√≠klad:
```gherkin
Feature: User login
  Scenario: Successful login
    Given user is on login page
    When user enters valid credentials
    Then user is redirected to dashboard
```

## 3. D≈Øle≈æit√© detaily

### ƒåast√© chyby a jak se jim vyhnout

- **Z√°vislosti mezi testy**: Testy by nemƒõly ovliv≈àovat sebe navz√°jem. Pou≈æ√≠vejte fixtures s `scope="function"` nebo `autouse`.
- **P≈ô√≠li≈° mnoho mockingu**: P≈ô√≠li≈°n√© pou≈æ√≠v√°n√≠ mock≈Ø m≈Ø≈æe v√©st k nedostateƒçn√©mu testov√°n√≠ re√°ln√©ho chov√°n√≠.
- **Nedostateƒçn√© pokryt√≠**: Vƒõt≈°inou se dƒõl√° v√≠ce jednotkov√Ωch test≈Ø, ale chyb√≠ integraƒçn√≠ nebo E2E.

### Best practices

- Testujte v≈ædy nejen spr√°vn√© v√Ωsledky, ale i edge case.
- Pou≈æ√≠vejte n√°zvy test≈Ø, kter√© jasnƒõ popisuj√≠, co se testuje.
- Dƒõlte testy do logick√Ωch skupin podle funkc√≠.

### Performance tipy

- Pou≈æ√≠vejte `pytest-xdist` pro paraleln√≠ spou≈°tƒõn√≠ test≈Ø.
- Vyu≈æ√≠vejte `pytest-cov` pro rychl√© z√≠sk√°n√≠ pokryt√≠ bez nutnosti spou≈°tƒõt v≈°echny testy znovu.

Z√°kladn√≠ pou≈æit√≠ `@pytest.mark.parametrize` a `@pytest.fixture` pro vytv√°≈ôen√≠ testovac√≠ch dat a p≈ôedp≈ôipraven√Ωch objekt≈Ø.

````
# test_example_1.py
import pytest

# Mockov√°n√≠ extern√≠ slu≈æby
class UserService:
    def get_user(self, user_id):
        # Simulace vol√°n√≠ API nebo datab√°ze
        if user_id == 1:
            return {"id": 1, "name": "Jan Nov√°k", "email": "jan@example.com"}
        elif user_id == 2:
            return {"id": 2, "name": "Petr Svoboda", "email": "petr@example.com"}
        else:
            return None

# Fixture pro vytvo≈ôen√≠ instance slu≈æby
@pytest.fixture
def user_service():
    """Vytvo≈ô√≠ instanci UserService pro testy"""
    return UserService()

# Parametrizovan√© testy s r≈Øzn√Ωmi daty
@pytest.mark.parametrize("user_id,expected_name", [
    (1, "Jan Nov√°k"),
    (2, "Petr Svoboda"),
    (3, None)
])
def test_get_user(user_service, user_id, expected_name):
    """Testuje z√≠sk√°n√≠ u≈æivatele podle ID"""
    user = user_service.get_user(user_id)
    
    if expected_name is None:
        assert user is None
    else:
        assert user["name"] == expected_name

# Test s fixture pou≈æit√≠m
def test_user_exists(user_service):
    """Testuje, ≈æe u≈æivatel existuje"""
    user = user_service.get_user(1)
    assert user is not None
    assert user["id"] == 1
    assert user["name"] == "Jan Nov√°k"
````

### P≈ô√≠klad 6

### Co demonstruje:
Pou≈æit√≠ `monkeypatch` pro nahrazen√≠ metod bƒõhem testu, ƒç√≠m≈æ se testuje chov√°n√≠ aplikace bez skuteƒçn√©ho vol√°n√≠ extern√≠ch slu≈æeb.

````
# test_example_2.py
import pytest
import requests

class APIClient:
    def __init__(self, base_url):
        self.base_url = base_url
    
    def get_data(self, endpoint):
        """Vol√° API a vrac√≠ data"""
        response = requests.get(f"{self.base_url}/{endpoint}")
        return response.json()

@pytest.fixture
def api_client():
    """Vytvo≈ô√≠ klienta s testovac√≠m URL"""
    return APIClient("https://api.example.com")

# Mockov√°n√≠ request.get pomoc√≠ monkeypatch
def test_get_data_with_mock(monkeypatch, api_client):
    """Testuje API vol√°n√≠ s mockovan√Ωm HTTP po≈æadavkem"""
    
    # Definujeme mockovou odpovƒõƒè
    mock_response = type('MockResponse', (), {
        'json': lambda: {'data': 'test_data', 'status': 'success'}
    })()
    
    # Nahrad√≠me requests.get mockem
    monkeypatch.setattr(requests, 'get', lambda url: mock_response)
    
    # Spust√≠me test
    result = api_client.get_data("users")
    
    # Ovƒõ≈ô√≠me v√Ωsledek
    assert result['data'] == 'test_data'
    assert result['status'] == 'success'

# Test s chybou
def test_get_data_with_error(monkeypatch, api_client):
    """Testuje chybov√© chov√°n√≠ API"""
    
    # Mockujeme chybu
    def mock_get(url):
        response = type('MockResponse', (), {
            'json': lambda: {'error': 'Server error'}
        })()
        return response
    
    monkeypatch.setattr(requests, 'get', mock_get)
    
    result = api_client.get_data("users")
    
    assert result['error'] == 'Server error'
````

### Co demonstruje:
Integraƒçn√≠ testy, kter√© testuj√≠ komunikaci mezi aplikac√≠ a datab√°z√≠.

````
# test_example_3.py
import pytest
import sqlite3
from contextlib import contextmanager

# Simulace modelu u≈æivatele
class User:
    def __init__(self, name, email):
        self.name = name
        self.email = email

# Modelov√° t≈ô√≠da pro pr√°ci s datab√°z√≠
class UserModel:
    def __init__(self, db_path=":memory:"):
        self.db_path = db_path
        self._create_table()
    
    def _create_table(self):
        """Vytvo≈ô√≠ tabulku u≈æivatel≈Ø"""
        conn = sqlite3.connect(self.db_path)
        cursor = conn.cursor()
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS users (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                name TEXT NOT NULL,
                email TEXT UNIQUE NOT NULL
            )
        ''')
        conn.commit()
        conn.close()
    
    def create_user(self, user):
        """Vytvo≈ô√≠ nov√©ho u≈æivatele"""
        conn = sqlite3.connect(self.db_path)
        cursor = conn.cursor()
        cursor.execute(
            "INSERT INTO users (name, email) VALUES (?, ?)",
            (user.name, user.email)
        )
        conn.commit()
        conn.close()
    
    def get_user_by_email(self, email):
        """Vr√°t√≠ u≈æivatele podle e-mailu"""
        conn = sqlite3.connect(self.db_path)
        cursor = conn.cursor()
        cursor.execute("SELECT * FROM users WHERE email=?", (email,))
        result = cursor.fetchone()
        conn.close()
        
        if result:
            return {"id": result[0], "name": result[1], "email": result[2]}
        return None

@pytest.fixture
def user_model():
    """Vytvo≈ô√≠ instanci modelu u≈æivatele s doƒçasnou datab√°z√≠"""
    model = UserModel()
    yield model
    
    # ƒåi≈°tƒõn√≠ po testu (voliteln√© pro in-memory db)

def test_create_and_get_user(user_model):
    """Testuje vytvo≈ôen√≠ a naƒçten√≠ u≈æivatele"""
    
    # Vytvo≈ô√≠me nov√©ho u≈æivatele
    user = User("Jan Nov√°k", "jan@example.com")
    user_model.create_user(user)
    
    # Naƒçteme u≈æivatele podle e-mailu
    found_user = user_model.get_user_by_email("jan@example.com")
    
    # Ovƒõ≈ô√≠me, ≈æe je spr√°vn√Ω
    assert found_user is not None
    assert found_user["name"] == "Jan Nov√°k"
    assert found_user["email"] == "jan@example.com"

def test_user_not_found(user_model):
    """Testuje p≈ô√≠pad, kdy u≈æivatel nen√≠ nalezen"""
    
    user = user_model.get_user_by_email("nonexistent@example.com")
    
    assert user is None

def test_duplicate_email(user_model):
    """Testuje vytvo≈ôen√≠ u≈æivatele s duplicitn√≠m e-mailem"""
    
    # Vytvo≈ô√≠me prvn√≠ho u≈æivatele
    user1 = User("Jan Nov√°k", "jan@example.com")
    user_model.create_user(user1)
    
    # Vytvo≈ô√≠me druh√©ho u≈æivatele se stejn√Ωm e-mailem
    user2 = User("Petr Svoboda", "jan@example.com")
    
    # Ovƒõ≈ô√≠me, ≈æe to zp≈Øsob√≠ chybu (unik√°tn√≠ constraint)
    with pytest.raises(sqlite3.IntegrityError):
        user_model.create_user(user2)
````
